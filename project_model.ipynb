```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso

from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_excel("data (1).xlsx")
```

```python
df.head()
```

# train-test split


```python
xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

# Ridge model


```python
ridge = Ridge(alpha=1.0)

ridge.fit(xtrain, ytrain)
```

prediction

```python
y_pred = ridge.predict(xtest)
```

```python
print("Ridge MSE:", mean_squared_error(ytest, y_pred))
print("Ridge R2:", r2_score(ytest, y_pred))
```

# Lasso model


```python
lasso = Lasso(alpha=0.1)
lasso.fit(xtrain, ytrain)
```

prediction

```python
y_pred_lasso = lasso.predict(xtest)
```

```python
print("Lasso MSE:", mean_squared_error(ytest, y_pred_lasso))
print("Lasso R2:", r2_score(ytest, y_pred_lasso))
```
